```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://companiesmarketcap.com/usa/largest-companies-in-the-usa-by-market-cap/"
```

```python
r=requests.get(url)
soup=BeautifulSoup(r.content,"html.parser")
soup
```

```python
print(soup.prettify())
```

```python
companies=soup.find_all("td",class_="name-td")

companies
```

```python
names=[]
for company in companies:
  names.append(company.get_text())

names
```

```python
print(names)
```

```python
details=soup.find_all("td",class_="td-right")
DATA=[]
for detail in details:
  DATA.append(detail.get_text())
DATA
```

```python

rank=DATA[::3]
```

```python
shareprice=DATA[2::3]
```

```python
marketcap=DATA[1::3]
```

```python
import pandas as pd
df=pd.DataFrame()
df['rank']=rank
df['company']=names
df['shareprice']=shareprice
df['marketcap']=marketcap
df
```

```python
df.to_csv('top100USAcompanies.csv',index=False)
```
